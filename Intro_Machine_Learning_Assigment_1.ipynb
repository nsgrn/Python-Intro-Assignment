#Step 14
Build your K-neighbors classifier for nearest neighbor of 1 using from sklearn.neighbors import KNeighborsClassifier, fit the model to your train dataset and make a prediction for teh data point of [5, 1.9, 1, 0.2]. Print your prediction class value as an integer and also the corresponding string label.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, Y_train)
new_sample = [[5, 1.9, 1, 0.2]]
prediction = knn.predict(new_sample)
print("Predicted class (integer):", prediction[0])
print("Predicted class (label):", iris_dataset.target_names[prediction[0]])
```

```python
def min_max_scaler(x):
  return (x - x.min()) / (x.max() - x.min())
```

```python
scaled_df = auto_df.copy()
scaled_disp = min_max_scaler(auto_df['displacement'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled_cols = ['displacement', 'horsepower', 'acceleration', 'weight', 'mpg']
my_scaler = MinMaxScaler()
df_scaled = auto_df.copy()
df_scaled[scaled_cols] = my_scaler.fit_transform(df_scaled[scaled_cols])
df_scaled.head()
```
