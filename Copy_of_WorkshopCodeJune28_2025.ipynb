```python
from langchain_google_genai import GoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt = PromptTemplate(
    input_variables = ['topic'],
    template = """ Explain {topic} in simple terms with an example:

    Format:
    - Keep t under 100 words
    - Use an analogy
    - End with a practical example

    Topic: {topic}
    Explanation: """
    )
filled_prompt = prompt.format(topic='AI agent')

print(filled_prompt)
```

```python
#first give the prompt,
#then connect to llm and get the response,
#then format the response and give it as the output


chain = prompt | llm | StrOutputParser()
```

```python
topics = ['blokchain', 'AI agent', 'Internet Model']

for t in topics:
  print(f'\n Explaining: {t}')

  result = chain.invoke({'topic': t})

  print(f'Response: {result}')
```

```python
from langchain.vectorstores import FAISS
from langchain.schema import Document
```

```python
documents
```

```python
docs = [Document(page_content=doc) for doc in documents]

docs
```

```python
vectorstore = FAISS.from_documents(docs,embeddings)
```

```python
vectorstore
```

```python
test_query = 'refund policy'

relevant_document = vectorstore.similarity_search(test_query, k=2)

print(f'{relevant_document[0].page_content} ')
```

```python
print(relevant_document)

'\n'.join([doc.page_content for doc in relevant_document])

```

```python
def ask_rag(question):
  relevant_document = vectorstore.similarity_search(question, k=2)

  context = '\n'.join([doc.page_content for doc in relevant_document])

  rag_prompt = f"""
Based on this company information:
{context}

Question: {question}

Please
- provide a helpful answer based on the company information provided.
- first greet the customer, say a few nice words

Answer:
  """

  response = llm.invoke(rag_prompt)
  return response

question = 'What is your return policy'

answer = ask_rag(question)

print(answer)
```
